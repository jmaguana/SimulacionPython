# Generación de números pseudoaleatorios
## Método congruencia lineal

```python
from tabulate import tabulate

a = 3
b = 5
m = 19
xn = 1

iteraciones = 15
tabla = []
tabla.append(["Iteracion","Xn","Ui"])
tabla.append([0,xn])
for i in range(1,iteraciones+1):
    numero = (a*xn+b)%m
    valor = numero/m
    tabla.append([i,numero, valor])
    xn = numero

print(tabulate(tabla, tablefmt='fancy_grid'))
```
